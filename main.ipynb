# Data cleaning

```python
df.sample(5)
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
df.sample(5)
```

```python
df.duplicated().sum()
```

```python
df.columns.to_list()
```

```python
# i think if col names is lower i is easier to work with
df.columns = df.columns.str.lower()
```

```python
df.sample(3)
```

```python
df.isna().sum()
```

```python
df.isna().mean()*100
```

```python
# virtually every column has 5 % missing NaN value, i think this is a very subset of the dateset, so it is i think need to be filled.
```

```python
df.dtypes
```

```python
df.columns.to_list()
```

```python
# median is good to fill with

floating_cols = [
 'sleep_hours',
 'study_hours',
 'social_media_hours',
 'attendance',
 'exam_pressure',
 'family_support',]

for col in floating_cols:
    df[col] = df[col].fillna(df[col].median())
print("Floating columns NaN values are filled with median of that columns.")
```

```python
df.isna().sum()
```

```python
# filling month with median is not good as its meadian is 6 7 month - june, july. Filling them with mode is better i think. 
df["month"].mode()
```

```python
df["month"] = df["month"].fillna(df["month"].median())
print("Month NaN values are filled with Mode(3.0 - March)")
```

```python
df.isna().sum()
```

```python
df.student_type.dtype
```

```python
df.student_type.value_counts()
```

```python
df['student_type'] = df['student_type'].fillna(df['student_type']).mode()[0]
print("Student_type NaN values are filled with Mode(college)")
```

```python
df.isna().sum(), df.dtypes
```

```python
df.sample(5)
```

```python
df.info()
```

```python
df.month.value_counts().sort_values()
```

```python
# i think month must be as its name like categorical not numbers, so it is better to move them to month
month_map = {
    1: 'January',   2: 'February',  3: 'March',
    4: 'April',     5: 'May',       6: 'June',
    7: 'July',      8: 'August',    9: 'September',
    10: 'October',  11: 'November', 12: 'December'
}
df['month'] = df['month'].astype(int).map(month_map)
print("Month numbers transferred to categories.")
```
